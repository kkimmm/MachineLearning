```python
#——————————————————example 1——————————————————————

import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from pylab import *

# 假设有如下测试点集：（3,1）（3,2）（4,1）（4,2）（1,3）（1,4）（2,3）（2,4）...使用KMeans算法对其进行聚类。
points = np.array([[3,1],[3,2],[4,1],[4,2],[1,3],[1,4],[2,3],[2,4],[2,1],[2,4],[2.3,3.6],[4,1.2],[2.1,3.3],[3.2,2.2],[3.8,2],[2.8,1.7]])
plt.scatter(points[:,0],points[:,1]) #scatter绘制散点图
plt.show()

km = KMeans(n_clusters = 2)#调用聚类方法，聚成2类
km.fit(points)

centers = km.cluster_centers_#生成聚类中心点
print('聚类中心点为：\n',centers,'\n')

labels = km.labels_ #测试点集每个数据点所属分组
print('测试集每个数据点被划分的分组为：\n',labels)

plt.scatter(points[:,0], points[:,1], c=labels)
plt.scatter(centers[:,0],centers[:,1], marker='*', s=100)
predict = np.array([[1.5,3], [3.5,3]])
label = km.predict(predict)
plt.scatter(predict[:,0], predict[:,1], c = label,marker='x',s=200)
plt.show()

'''
#下面是另外一种数据显示方法
for i in range(len(labels)):
    plt.scatter(points[i][0], points[i][1], c=('r' if labels[i] == 0 else 'b'))
    plt.scatter(centers[:,0],centers[:,1], marker='*', s=100)
predict = [[1.5,3], [3.5,3]]# 预测
label = km.predict(predict)
for i in range(len(label)):
    plt.scatter(predict[i][0], predict[i][1], c=('r' if label[i] == 0 else 'b'), marker='x',s=200)
plt.show()
'''
```

```python
#———————————————————————example 2——————————————————————
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
'''
make_blobs函数是为聚类产生数据集
产生一个数据集和相应的标签
n_samples:表示数据样本点个数,默认值100
n_features:表示数据的维度，默认值是2
centers:产生数据的中心点，默认值3
cluster_std：数据集的标准差，浮点数或者浮点数序列，默认值1.0
center_box：中心确定之后的数据边界，默认值(-10.0, 10.0)
shuffle ：洗乱，默认值是True
random_state:官网解释是随机生成器的种子
更多参数即使请参考：http://scikit-learn.org/dev/modules/generated/sklearn.datasets.make_blobs.html#sklearn.datasets.make_blobs
'''
# X为样本特征，y为样本簇类别， 共1000个样本，每个样本4个特征，共4个簇，簇中心在[-1,-1], [0,0],[1,1], [2,2]， 簇方差分别为[0.4, 0.2, 0.2]
X, y = make_blobs(n_samples=1000, n_features=2, centers=[[-1,-1], [0,0], [1,1], [2,2]], cluster_std=[0.4, 0.2, 0.2, 0.2], 
                  random_state =9)
plt.scatter(X[:, 0], X[:, 1], marker='o')
plt.show()
print("样本被划分的类别分别为：\n",y)

from sklearn.cluster import KMeans
y_pred = KMeans(n_clusters=4).fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.title("cluster")
plt.show()

from sklearn import metrics
metrics.calinski_harabaz_score(X, y_pred)  #Calinski-Harabasz Index评估的聚类分数
```

```python
#———————————————————————example 3——————————————————————

import numpy as np
from sklearn.cluster import KMeans
 
def loadData(filePath):
    fr = open(filePath,'r+')
    lines = fr.readlines()
    retData = []
    retCityName = []
    for line in lines:
        items = line.strip().split(",")#strip()方法用于移除字符串头、尾指定的字符（默认为空格）
                                       #split()通过指定分隔符对字符串进行切片，如果参数num 有指定值，则仅分隔 num 个子字符串
        retCityName.append(items[0])
        retData.append([float(items[i]) for i in range(1,len(items))])
    return retData,retCityName

if __name__ == '__main__':
    data,cityName = loadData('/Users/gkai/Documents//city.txt')
    km = KMeans(n_clusters=3)
    label = km.fit_predict(data)
    expenses = np.sum(km.cluster_centers_,axis=1)#expense:聚类中心点的数值加和，也就是平均消费水平
    #print(expenses)
    CityCluster = [[],[],[]]
    for i in range(len(cityName)):
        CityCluster[label[i]].append(cityName[i])
    for i in range(len(CityCluster)):
        print("Expenses:%.2f" % expenses[i])
        print(CityCluster[i])
```
