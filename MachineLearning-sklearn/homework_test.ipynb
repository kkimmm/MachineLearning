```python
from sklearn.cluster import KMeans
from sklearn import preprocessing

X_scaled = preprocessing.scale(data)#均值归一化（按列均值归一化）

X_scaled = X_scaled.T
from sklearn.decomposition import PCA #pca降维
pca = PCA(n_components=10)

pcadata = pca.fit(X_scaled).transform(X_scaled)

km = KMeans(n_clusters = 5)#调用聚类方法，聚成n_clusters种类型
km.fit(pcadata)
labels = km.labels_ #测试点集每个数据点所属分组
print('测试集每个数据点被划分的分组为：\n',labels)

DataFrame(pcadata)
#plt.plot(DataFrame(labels))
#plt.show()
```
